```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/home/marcius/mb_files/19.python/ssd/stall_IF/if_optuna_stall.csv')
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
sys_filter = 'system_2'
dt=df[df['system'] == sys_filter]

fig = go.Figure()
fig.add_trace(go.Scatter(x=dt['timestamp'], y=dt.loc[:,'steam'],mode='lines',name='T1'))
fig.add_trace(go.Scatter(x=dt['timestamp'], y=dt.loc[:,'condensate'],mode='lines',name='T2'))
fig.add_trace(go.Scatter(x=dt['timestamp'], y=dt.loc[:,'stall_severity']*10, mode='lines',name='IF Optuna - stall'))

fig.update_layout(title='Temperature profiles', xaxis_title='timestamp',yaxis_title='Temperature [C]')
fig.show()
```
